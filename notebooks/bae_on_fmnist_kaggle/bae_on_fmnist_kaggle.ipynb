```python
import numpy as np
from experiments.bae_on_the_two_nist import filter_iterations
from experiments.bae_on_the_two_nist_cnn import run_trainings
from hpc.utils import make_pairs_named_tuple
```

# Common Parameters

```python
latent_dimensions = [0]
learning_rates = [0.001]
seeds = np.arange(0, 20, step=1)
directory = "data/bae_on_fmnist_kaggle_ae"
dataset = ["fmnist"]
noise_factors = [0, 0.25,0.5, 0.75]
epochs = 20
```

# UAE

```python
compress_loss_iterations_data = make_pairs_named_tuple(
    loss=["compression_loss"],
    noise_factor=noise_factors,
    latent_dim=latent_dimensions,
    learning_rate=learning_rates,
    seed=seeds,
    dataset_name=dataset,
    directory=[directory]
)
```

# BAE

```python
decompr_factors = [0.9]
n_factors = [0.1]
training_types = [0,1,2]
```

```python
benign_loss_iterations_data = make_pairs_named_tuple(
    loss=["benign_loss"],
    noise_factor=noise_factors,
    d_weight=decompr_factors,
    n_weight=n_factors,
    latent_dim=latent_dimensions,
    learning_rate=learning_rates,
    training_type=training_types,
    seed=seeds,
    dataset_name=dataset,
    directory=[directory]
)
```
